```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define the confusion matrix data
labels = ['calm', 'dancing', 'energetic', 'happy', 'neutral', 'relaxing', 'sad']
confusion_matrix = np.array([
    [10640, 0, 1, 0, 6, 0, 0],
    [0, 4809, 0, 0, 237, 0, 0],
    [55, 0, 23866, 0, 643, 0, 0],
    [0, 0, 0, 30853, 0, 0, 5143],
    [8, 2927, 8, 0, 60116, 0, 0],
    [320, 0, 0, 0, 840, 28960, 0],
    [0, 0, 0, 5814, 0, 0, 8097]
])

# Create the plot
plt.figure(figsize=(8,6))

# Create heatmap using seaborn
sns.heatmap(confusion_matrix,
            annot=True,  # Show numbers in cells
            fmt='d',     # Use integer format
            cmap='YlOrRd',  # Yellow-Orange-Red colormap
            xticklabels=labels,
            yticklabels=labels,
            cbar_kws={'label': ''})

# Customize the plot
plt.title('Confusion Matrix', pad=20)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

# Rotate x-axis labels for better readability
plt.xticks(rotation=0)
plt.yticks(rotation=0)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()

# Optional: Calculate and print some metrics
total_predictions = confusion_matrix.sum()
correct_predictions = np.trace(confusion_matrix)
accuracy = correct_predictions / total_predictions

print(f"\nModel Performance Metrics:")
print(f"Total Predictions: {total_predictions:,}")
print(f"Correct Predictions: {correct_predictions:,}")
print(f"Overall Accuracy: {accuracy:.2%}")

# Calculate per-class metrics
for i, label in enumerate(labels):
    true_positives = confusion_matrix[i, i]
    total_actual = confusion_matrix[i, :].sum()
    total_predicted = confusion_matrix[:, i].sum()
    
    precision = true_positives / total_predicted if total_predicted > 0 else 0
    recall = true_positives / total_actual if total_actual > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    print(f"\n{label.capitalize()} Class Metrics:")
    print(f"Precision: {precision:.2%}")
    print(f"Recall: {recall:.2%}")
    print(f"F1-Score: {f1:.2%}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Data
categories = ['calm', 'dancing', 'energetic', 'happy', 'neutral', 'relaxing', 'sad']
precision = [0.97, 0.74, 1.00, 0.81, 0.96, 1.00, 0.70]
recall = [1.00, 0.78, 0.97, 0.92, 0.98, 0.96, 0.45]
f1_score = [0.98, 0.76, 0.99, 0.87, 0.97, 0.98, 0.55]

# Create custom YlOrRd colormap
colors = ['#fed976', '#feb24c', '#fd8d3c', '#fc4e2a', '#e31a1c', '#b10026']
n_bins = 3  # We need 3 colors for our 3 metrics
ylord = LinearSegmentedColormap.from_list('custom_ylord', colors)
colors_for_bars = [ylord(i) for i in np.linspace(0, 0.8, n_bins)]

# Set up the figure
plt.figure(figsize=(8, 6))

# Set the width of each bar and positions of the bars
width = 0.25
x = np.arange(len(categories))

# Create bars with YlOrRd colors
bars1 = plt.bar(x - width, precision, width, label='Precision', color=colors_for_bars[0], alpha=0.9)
bars2 = plt.bar(x, recall, width, label='Recall', color=colors_for_bars[1], alpha=0.9)
bars3 = plt.bar(x + width, f1_score, width, label='F1-score', color=colors_for_bars[2], alpha=0.9)

# Customize the plot
plt.title('Random Forest Model Performance Metrics', pad=20, size=14)
plt.xlabel('Categories', size=12)
plt.ylabel('Score', size=12)

# Set the x-axis labels
plt.xticks(x, categories, rotation=45)

# Add gridlines for better readability
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# Set y-axis limits from 0 to 1
plt.ylim(0, 1.1)

# Add value labels on top of each bar
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.2f}',
                ha='center', va='bottom')



# Add legend
plt.legend(loc='lower left')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

# Print summary statistics
print("\nAverage Metrics:")
print(f"Average Precision: {np.mean(precision):.3f}")
print(f"Average Recall: {np.mean(recall):.3f}")
print(f"Average F1-Score: {np.mean(f1_score):.3f}")

print("\nBest Performing Categories:")
print(f"Best Precision: {categories[np.argmax(precision)]} ({max(precision):.3f})")
print(f"Best Recall: {categories[np.argmax(recall)]} ({max(recall):.3f})")
print(f"Best F1-Score: {categories[np.argmax(f1_score)]} ({max(f1_score):.3f})")

print("\nWorst Performing Categories:")
print(f"Worst Precision: {categories[np.argmin(precision)]} ({min(precision):.3f})")
print(f"Worst Recall: {categories[np.argmin(recall)]} ({min(recall):.3f})")
print(f"Worst F1-Score: {categories[np.argmin(f1_score)]} ({min(f1_score):.3f})")
```
